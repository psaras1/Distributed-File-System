Most variables are **indices** — they compress multiple things into a single score.

## Reservations

Countries can sign treaties and still limit what they accept.  
That’s what reservations are.

- `iccpr_reservation`, `cedaw_reservation`, `cat_reservation`  
  → 1 = has at least one reservation  

- `total_reservations`  
  → number of reservations  

- `any_reservations`  
  → 1 = at least one reservation somewhere  

You can think of reservations as:

> “We agree — but not fully.”

```python
# use this to write code, remember you can add more if you need by pressing '+' Code
# bucket raw numbers -> readable categories

def categorise(x):
  if x == 0: return "0"
  elif x == 1: return "1"
  elif x == 2: return "2"
  else: return "3+"

# take function we defined, apply it to every row in total_reservations, store in new column
df["reservation_group"] = df["total_reservations"].map(categorise)


# count countries per group, sort to make bars go 0,1,2,3+ left to right
# value_counts count how many times each value appears
# sort_index sorts by index(labels in this case)
df["reservation_group"].value_counts().sort_index().plot(kind="bar", title="Res Distribution")
plt.xlabel("Reservations")
plt.ylabel("Countries")
plt.show()
```
